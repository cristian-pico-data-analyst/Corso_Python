### Primo grafico a barre (MatplotLib)

Creazione del grafico

```python
import matplotlib.pylab as plt

# Dati: corsi e numero di studenti iscritti
corsi = ['Python', 'Data Science', 'Web Design', 'AI Basics', 'Cybersecurity']
studenti = [25, 18, 30, 22, 15]

# Creazione del grafico a barre
plt.bar(corsi, studenti) # crea un grafico a barre con i corsi sull'asse x e il numero di studenti sull'asse y, con barre di colore azzurro

plt.show() # visualizza il grafico a barre creato
```

Personalizzare etichette e titolo

```python
# Dati: corsi e numero di studenti iscritti
corsi = ['Python', 'Data Science', 'Web Design', 'AI Basics', 'Cybersecurity']
studenti = [25, 18, 30, 22, 15]

# Creazione del grafico a barre
plt.bar(corsi, studenti, color='skyblue') # crea un grafico a barre con i corsi sull'asse x e il numero di studenti sull'asse y, con barre di colore azzurro
plt.xlabel('Corsi') # aggiunge l'etichetta "Corsi" all'asse x
plt.ylabel('Numero di Studenti') # aggiunge l'etichetta "Numero di Studenti" all'asse y
plt.title('Iscrizioni ai Corsi') # aggiunge il titolo "Iscrizioni ai Corsi" al grafico

plt.show()
```

### Secondo grafico a barre Raggruppate (MatplotLib)

##### Utile per confrontare i voti medi di due elementi diversi (Classe A o Classe B).

Inseriamo due barre affiancate

```python
import matplotlib.pylab as plt # importa la libreria matplotlib per poter creare grafici a partire dai dataframe
import numpy as np # importa la libreria numpy per poter creare array di dati numerici

corsi = ['Matematica', 'Fisica', 'Informatica', 'Chimica', 'Biologia', 'Alimentazione', 'Nutrizione', 'Scienze Motorie', 'Psicologia', 'Sociologia', 'Italiano', 'Storia', 'Geografia', 'Arte', 'Musica', 'Inglese', 'Spagnolo', 'Francese', 'Inglese']
voti_A = [28.8, 27.7, 27, 25.6, 29, 26, 28, 24, 30, 27.6, 29, 25.6, 28.6, 30, 26, 29, 27.5, 25, 29 ]
voti_B = [24, 26.3, 22, 20.5, 27, 25, 27.3, 23, 29, 26, 28.5, 24, 27, 29, 25, 28, 26.2, 29.1, 27]

# Posizioni delle etichette
x = np.arange(len(corsi)) # crea un array di posizioni per le etichette dei corsi sull'asse x
width = 0.35 # larghezza delle barre

# Confronta le due classi con barre affiancate
fig, ax = plt.subplots() # crea una figura e un asse per il grafico
rects1 = ax.bar(x - width/2, voti_A, width, label='Classe A', color='lightblue') # crea le barre per la classe A, posizionandole a sinistra delle posizioni x
rects2 = ax.bar(x + width/2, voti_B, width, label='Classe B', color='salmon') # crea le barre per la classe B, posizionandole a destra delle posizioni x

plt.show()
```

Personalizzo il grafico con titolo, label, etichette e legenda

```python
import matplotlib.pylab as plt # importa la libreria matplotlib per poter creare grafici a partire dai dataframe
import numpy as np # importa la libreria numpy per poter creare array di dati numerici

corsi = ['Matematica', 'Fisica', 'Informatica', 'Chimica', 'Biologia', 'Alimentazione', 'Nutrizione', 'Scienze Motorie', 'Psicologia', 'Sociologia', 'Italiano', 'Storia', 'Geografia', 'Arte', 'Musica', 'Inglese', 'Spagnolo', 'Francese', 'Inglese']
voti_A = [28.8, 27.7, 27, 25.6, 29, 26, 28, 24, 30, 27.6, 29, 25.6, 28.6, 30, 26, 29, 27.5, 25, 29]
voti_B = [24, 26.3, 22, 20.5, 27, 25, 27.3, 23, 29, 26, 28.5, 24, 27, 29, 25, 28, 26.2, 29.1, 27]

# Posizioni delle etichette
x = np.arange(len(corsi)) # crea un array di posizioni per le etichette dei corsi sull'asse x 
width = 0.35 # larghezza delle barre
fig, ax = plt.subplots() # crea una figura e un asse per il grafico
rects1 = ax.bar(x - width/2, voti_A, width, label='Classe A', color='lightblue') # crea le barre per la classe A, posizionandole a sinistra delle posizioni x
rects2 = ax.bar(x + width/2, voti_B, width, label='Classe B', color='salmon') # crea le barre per la classe B, posizionandole a destra delle posizioni x

# Personalizza il grafico con etichette e titolo
ax.set_xlabel('Corsi') # imposta l'etichetta dell'asse x
ax.set_ylabel('Media Voti') # imposta l'etichetta dell'asse y 
ax.set_title('Confronto voti tra classi') # imposta il titolo del grafico

# Imposta le etichette dei corsi sull'asse x
ax.set_xticks(x) # imposta le posizioni delle etichette sull'asse x
ax.set_xticklabels(corsi, rotation=45, ha='right') # imposta le etichette dei corsi sull'asse x
ax.legend() # mostra la legenda del grafico

plt.show()
```

### Heatmap (Seaborn)

- Utile per visualizzare le presenze degli studenti durante la settimana

```python
import seaborn as sns # importa la libreria seaborn per poter creare grafici più avanzati a partire dai dataframe
import pandas as pd # importa la libreria pandas per poter manipolare i dati in formato tabellare
import matplotlib.pyplot as plt # importa la libreria matplotlib per poter visualizzare i grafici creati con seaborn

# Dati: Righe studenti, colonne = giorni della settimana (1 = Presente, 0 = Assente)
dati = {
    'Lun': [1, 0, 1, 1, 0, 1, 0],
    'Mar': [1, 1, 0, 1, 1, 0, 1],
    'Mer': [0, 1, 1, 0, 1, 1, 0],
    'Gio': [1, 0, 1, 1, 0, 1, 0],
    'Ven': [0, 1, 0, 1, 1, 0, 1]
}

studenti = ["Cristian", "Giovanni", "Maria", "Mia", "Flavia", "Luca", "Romeo"]
df = pd.DataFrame(dati, index=studenti) # crea un dataframe con i dati di presenza, usando i nomi degli studenti come indice

sns.heatmap(df, annot=True, cmap='YlGnBu') # crea una heatmap a partire dal dataframe, con annotazioni dei valori e una mappa di colori "YlGnBu"

plt.show() # visualizza la heatmap creata
```
