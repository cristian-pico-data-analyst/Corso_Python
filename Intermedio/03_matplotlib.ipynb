### Heatmap (Seaborn)

```python
import seaborn as sns # importa la libreria seaborn per poter creare grafici più avanzati a partire dai dataframe
import pandas as pd # importa la libreria pandas per poter manipolare i dati in formato tabellare
import matplotlib.pyplot as plt # importa la libreria matplotlib per poter visualizzare i grafici creati con seaborn

# Dati: Righe studenti, colonne = giorni della settimana (1 = Presente, 0 = Assente)
dati = {
    'Lun': [1, 0, 1, 1, 0, 1, 0],
    'Mar': [1, 1, 0, 1, 1, 0, 1],
    'Mer': [0, 1, 1, 0, 1, 1, 0],
    'Gio': [1, 0, 1, 1, 0, 1, 0],
    'Ven': [0, 1, 0, 1, 1, 0, 1]
}

studenti = ["Cristian", "Giovanni", "Maria", "Mia", "Flavia", "Luca", "Romeo"]
df = pd.DataFrame(dati, index=studenti) # crea un dataframe con i dati di presenza, usando i nomi degli studenti come indice

sns.heatmap(df, annot=True, cmap='YlGnBu') # crea una heatmap a partire dal dataframe, con annotazioni dei valori e una mappa di colori "YlGnBu"
plt.title('Registro Presenze degli Studenti') # aggiunge un titolo alla heatmap

plt.show() # visualizza la heatmap creata
```

### Grafico Radar (Matplotlib + Numpy)

Utile per: valutare le competenze di un singolo studente in diverse aree.

```python
import matplotlib.pyplot as plt
import numpy as np

# Creiamo una lista competenze
labels = np.array(['Coding', 'Teoria', 'Progetto', 'Data Analyst', 'Reverse engineering', 'Sviluppo', 'Testing', 'Documentazione']) # crea un array di etichette per le competenze elencate

# Valori per uno studente
stats = np.array([8, 7, 9, 6, 5, 8, 7, 6]) # valori di competenza per uno studente in ciascuna delle competenze elencate in "labels"
"""
LOGICA
np.linspace(0, 2*np.pi, len(labels): crea un array di angoli equidistanti tra 0 e 2*pi, con un numero di elementi pari al numero di etichette nelle competenze
endpoint=False: esclude il punto finale 2*pi dall'array di angoli, in modo che le etichette siano posizionate in modo circolare senza sovrapporsi
.tolist(): converte l'array di angoli in una lista, che può essere utilizzata per posizionare le etichette in un grafico radar o a ragnatela
"""
angles = np.linspace(0, 2*np.pi, len(labels), endpoint=False).tolist() # crea un array di angoli per posizionare le etichette in modo circolare
angles += angles[:1] # aggiunge il primo angolo alla fine dell'array di angoli per chiudere il grafico radar
stats = np.concatenate((stats, [stats[0]])) # aggiunge il primo valore di competenza alla fine dell'array di valori per chiudere il grafico radar

""" IMPOSTAZIONE GRAFICO RADAR """
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True)) # crea una figura di dimensioni 8x8 pollici e un subplot con coordinate polari per il grafico radar
ax.plot(angles, stats, color='skyblue', linewidth=2, linestyle='solid') # traccia una linea che collega i punti del grafico radar, con colore skyblue, spessore di 2 e stile solido
ax.fill(angles, stats, color='skyblue', alpha=0.20) # riempie l'area sotto la linea del grafico radar con colore skyblue e trasparenza del 20%
ax.set_xticks(angles[:-1]) # imposta le posizioni delle etichette sui punti del grafico radar, escludendo l'ultimo punto che è una ripetizione del primo
ax.set_xticklabels(labels, size=8) # imposta le etichette delle competenze sui punti del grafico radar
ax.set_title('Competenze dello Studente', size=25, color='blue') # aggiunge un titolo al grafico radar, una grandezza di 25 e colore blu

plt.show() # visualizza il grafico radar creato
```
