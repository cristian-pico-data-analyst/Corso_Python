# IMPORTARE DATAFRAME

```python
import pandas as pd
studenti = pd.read_csv("studenti.csv") # carica il file CSV "studenti.csv" in un dataframe chiamato "studenti"
corso = pd.read_csv("corso.csv") # carica il file CSV "corso.csv" in un dataframe chiamato "corso"
```

# SALVARE IN CSV

Salvare in formato .csv il dataframe corso

```python
corso.to_csv("corso.csv", index=False) # salva il dataframe "corso" in un file CSV chiamato "corso.csv" senza includere l'indice
```

Salvare in formato .csv il dataframe studenti

```python
studenti.to_csv("studenti.csv", index=False) # salva il dataframe "studenti" in un file CSV chiamato "studenti.csv" senza includere l'indice
```
