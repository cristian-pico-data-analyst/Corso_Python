## Primo passo con pandas

### Crea un DataFrame = Dictionary = Dataset

```python
import pandas as pd

# Abbiamo creato un dataframe(dati) = dataset
data = {
    "Nome": ["Filippo", "Salomon", "Asia", "Ana", "Cristian", "Miriam"], 
    "Cognome": ["Battaglia", "Rossi", "Verdi", "Neri", "Gialli", "Blu"],
    "Voto": [8, 10, 9, 7, 11, 9]
    }

df = pd.DataFrame(data) # Lo trasformiamo in un dataframe
```

#### df visualizziamo il dataframe formato come tabella Excel

```python
df
```

#### print(df) si visualizza il dataframe come testo, simile ad un file CSV

```python
print(df)
```

#### df.head(1) # visualizza solo la prima riga del dataframe

```python
df.head(1) # visualizza solo la prima riga del dataframe
```

#### df.tail(2) # visualizza le ultime 2 righe del dataframe

```python
df.tail(2) # visualizza solo le ultime 2 righe del dataframe
```

#### df.info() # ci da informazioni sul dataframe, come il numero di righe, colonne, tipi di dati, ecc.

```python
df.info() # visualizza informazioni sul dataframe, come il numero di righe, colonne e tipi di dati
```

#### df.describe() # ci da informazioni statistiche sul dataframe, come la media, la deviazione standard, i valori minimi e massimi, ecc. (solo per colonne numeriche)

```python
df.describe() # visualizza statistiche descrittive del dataframe, come la media, la deviazione standard e i quartili per le colonne numeriche
```


#### df[["Nome","Cognome", "Età"]] # visualizza solo le colonne "Nome", "Cognome" e "Età" del dataframe
#### df["Età"] # visualizza solo la colonna "Età" del dataframe

```python
df["Nome"] # visualizza solo la colonna "Nome" del dataframe
```

Filtrare i dati per età

```python
df[df["Voto"] > 8] # visualizza solo le righe del dataframe in cui il voto è maggiore di 8
```

Calcolare media

```python
df["Voto"].mean() # calcola la media della colonna "Voto"
```

Aggiornare una colonna

```python
df["Promosso"] = df["Voto"] > 8 # crea una nuova colonna "Promosso" che indica se il voto è maggiore di 8
df
```

Mettere in ordine decrescente

```python
df.sort_values(by="Voto", ascending=False, inplace=True) # ordina il dataframe in base alla colonna "Voto" in ordine decrescente
df
```

Mettere in ordine crescente

```python
df.sort_values(by="Voto", ascending=True, inplace=True) # ordina il dataframe in base alla colonna "Voto" in ordine crescente
df
```
